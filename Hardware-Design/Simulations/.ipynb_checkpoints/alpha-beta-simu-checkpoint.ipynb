## Simulate ALPHA-BETA-GAMMA Estimattion Algorithm

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def alpha_beta(initial_estimate, measurements, units = None,true_value = None,plot = False):
    N = len(measurements); #Gets the number of measurements obtained.abs
    predicted = initial_estimate; #predict the next output using dynamic model
    outputs = np.zeros(N); #keeps the outputs
    current_estimate = 0;
    for i in range (N):
        gain = 1/(i+1) #compute the weighting coefficient
        current_estimate = predicted + gain *(measurements[i] - predicted) #state update equation
        outputs[i] = current_estimate;
        predicted = current_estimate; #using dynamic model
    if(plot):
        plt.plot(outputs, label = "estimates")
        plt.plot(measurements, label = "measurements")
        if true_value is  not None:
            true_values = true_value * np.ones(N);
            plt.plot(true_values, label ="true values");
            error = true_values - outputs
            plt.plot(error, label = "errors");
        plt.legend()
        plt.xlabel("Iterations");
        if units is not None:
            plt.ylabel(units);
        else:
            plt.ylabel("amplitudes")
    return current_estimate, predicted, outputs
        
```

```python
true_value = 100;
initial_guess = 90;
N = 10
measurements = np.random.randint(90,121,N) + np.random.randint(-5,6,N)
```

```python
_, _, outputs = alpha_beta(initial_guess,measurements,plot=True, units = "weight (m)")
```
