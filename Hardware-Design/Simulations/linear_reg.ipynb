```python
#imports

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import SGDRegressor, LinearRegression, Lasso, Ridge
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import PolynomialFeatures
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
weights = np.array ([30,63.7,44.3,52.2,37.7,34.3,47.3,60.1,56.7,36.6,32.2,42.3,55.3,54.5,44.7,56.6,54.4,45.5,52.8])
voltages = np.array([1.21,2.69,1.99,2.31,1.7,1.48,2.22,2.68,2.42,1.64,1.34,2.01,2.51,2.41,2.17,2.86,2.38,2.17,2.37])
X_train,X_test,y_train,y_test = train_test_split(weights,voltages,test_size=0.3)
len(weights)
```

## Training the Model

```python
simple_reg = LinearRegression()
simple_reg.fit(weights.reshape(-1,1),voltages);
simple_reg.coef_
```

```python
y_train_predict = simple_reg.predict(weights.reshape(-1,1))
```

```python
plt.scatter(weights,voltages, color = "c", label = "test")
plt.plot(weights,y_train_predict,color = "k", label = "predicted")
plt.legend()
plt.title("Linear Regression for weight-voltage relationship")
plt.ylabel("voltage")
plt.xlabel("weight")
```

```python
MSE_Test = mean_squared_error(voltages,y_train_predict)

print("MSE: {}".format(MSE_Test))
```

## Testing

```python
y_test_predict = simple_reg.predict(X_test.reshape(-1,1))
```

```python
MSE_Train = mean_squared_error(y_test,y_test_predict)
print("Train MSE: {}".format(MSE_Train))
```
